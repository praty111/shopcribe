```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d') # Change the format to match the actual date format (YYYY-MM-DD)
df['Delivery Date'] = pd.to_datetime(df['Delivery Date'])
df['Cancellation Date'] = pd.to_datetime(df['Cancellation Date'])
df.to_csv('data.csv', index=False)
```

```python
df['Total Order Value'] = pd.to_numeric(df['Total Order Value'], errors='coerce')
df['Shipping Charges'] = pd.to_numeric(df['Shipping Charges'], errors='coerce')
df.to_csv('data.csv', index=False)
```

```python
df.drop_duplicates(inplace=True)
df.to_csv('data.csv', index=False)
```

```python
df.info()
```

```python
#Delivery Time: Calculate the delivery time as the difference between Order Date and Delivery Date
df['Delivery Time'] = (df['Delivery Date'] - df['Order Date']).dt.days
df.to_csv('data.csv', index=False)
```

```python
#Discount Percentage: Calculate the discount percentage
df['Discount Percentage'] = (df['MRP'] - df['Discount Price']) / df['MRP'] * 100
df.to_csv('data.csv', index=False)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)  # Ensure Order Date is parsed correctly
df['Delivery Lead Time'] = (df['Delivery Date'] - df['Order Date']).dt.days
df.to_csv('data.csv', index=False)
#Delivery Lead Time: The number of days between the order date and the delivery date. This can help in understanding the efficiency of the delivery process.
```

```python
df['Cancellation Lead Time'] = (df['Cancellation Date'] - df['Order Date']).dt.days
df.to_csv('data.csv', index=False)
#Cancellation Lead Time: The number of days between the order date and the cancellation date. This feature might help in understanding customer behavior related to cancellations.
```

```python
df['Is Late Delivery'] = df['Delivery Lead Time'] > 7
df.to_csv('data.csv', index=False)
#Is Late Delivery: A binary feature indicating whether the delivery was made after a certain threshold (e.g., more than 7 days).
```

```python
df['Time Since Cancellation'] = (pd.Timestamp('today') - df['Cancellation Date']).dt.days
df.to_csv('data.csv', index=False)
#Time Since Cancellation: How long it has been since the order was canceled. This might be useful if you are looking at the impact of cancellations over time.
```

```python
df['Order Day of Week'] = df['Order Date'].dt.dayofweek
df['Delivery Day of Week'] = df['Delivery Date'].dt.dayofweek
df['Cancellation Day of Week'] = df['Cancellation Date'].dt.dayofweek
df.to_csv('data.csv', index=False)
#Day of Week: The day of the week on which the order was placed, delivered, or canceled. This can help in understanding trends related to specific days.
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Delivery Month'] = df['Delivery Date'].dt.month
df['Delivery Year'] = df['Delivery Date'].dt.year
df['Cancellation Month'] = df['Cancellation Date'].dt.month
df['Cancellation Year'] = df['Cancellation Date'].dt.year
df.to_csv('data.csv', index=False)
#Month and Year: The month and year of the order, delivery, or cancellation can capture seasonal trends.
```

```python
df['Previous Orders Count'] = df.groupby('Customer ID').cumcount()
df.to_csv('data.csv', index=False)
#Previous Orders Count: The number of orders a customer has placed before a specific order. This can help in understanding customer loyalty.
```

```python
df['Previous Cancellations Count'] = df.groupby('Customer ID')['Cancellation Date'].transform(lambda x: x.notnull().cumsum() - 1)
#Previous Cancellations Count: The number of previous orders that were canceled by the customer.
# Use transform instead of apply to return a Series with the same index as the original DataFrame.
df.to_csv('data.csv', index=False)
```

```python
df['Order_Delivery_Interaction'] = df['Order Month'] * df['Delivery Month']
df.to_csv('data.csv', index=False)
#Order-Delivery Interaction: The interaction between the time of order and delivery might reveal some insights.
```

```python
def get_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
df['Order Period'] = df['Order Date'].dt.hour.apply(get_period)
df.to_csv('data.csv', index=False)
#Order Period: Categorizing the order times into periods like "Morning", "Afternoon", "Evening", and "Night" based on the hour of the day.
```

```python
df['Total Orders'] = df.groupby('Customer ID')['Order ID'].transform('count')
df['Cancellation Rate'] = df.groupby('Customer ID')['Cancellation Date'].transform('count') / df['Total Orders']
df.to_csv('data.csv', index=False)
#Cancellation Rate: The ratio of canceled orders to total orders by a customer.

```

```python
import pandas as pd

# Sample data loading
# df = pd.read_csv('your_data.csv')

# Compute average order value per state
avg_order_value_state = df.groupby('State')['Total Order Value'].mean().reset_index()
avg_order_value_state.columns = ['State', 'Avg_Order_Value_State']

# Compute average delivery time per state
avg_delivery_time_state = df.groupby('State')['Delivery Time'].mean().reset_index()
avg_delivery_time_state.columns = ['State', 'Avg_Delivery_Time_State']

# Merge the aggregated features with the original dataframe
df = pd.merge(df, avg_order_value_state, on='State', how='left')
df = pd.merge(df, avg_delivery_time_state, on='State', how='left')

# Compute average order value per city
avg_order_value_city = df.groupby('City')['Total Order Value'].mean().reset_index()
avg_order_value_city.columns = ['City', 'Avg_Order_Value_City']

# Compute average delivery time per city
avg_delivery_time_city = df.groupby('City')['Delivery Time'].mean().reset_index()
avg_delivery_time_city.columns = ['City', 'Avg_Delivery_Time_City']

# Merge the aggregated features with the original dataframe
df = pd.merge(df, avg_order_value_city, on='City', how='left')
df = pd.merge(df, avg_delivery_time_city, on='City', how='left')
df.to_csv('data.csv', index=False)

# Display the updated dataframe
print(df.head())
```

```python
# Create a flag for discounted orders
df['Discounted_Order'] = df['Discount Price'] < df['MRP']
df.to_csv('data.csv', index=False)
```

```python
# Calculate total number of discounted and non-discounted orders
discounted_orders = df[df['Discounted_Order']].shape[0]
non_discounted_orders = df[~df['Discounted_Order']].shape[0]

# Ratio of discounted to non-discounted orders
discount_ratio = discounted_orders / (non_discounted_orders + 1)  # Adding 1 to avoid division by zero
df.to_csv('data.csv', index=False)
```

```python
# Calculate average discount percentage for discounted orders
avg_discount_percentage = df[df['Discounted_Order']]['Discount Percentage'].mean()
df.to_csv('data.csv', index=False)
```

```python
# Calculate average order value for discounted and non-discounted orders
avg_order_value_discounted = df[df['Discounted_Order']]['Total Order Value'].mean()
avg_order_value_non_discounted = df[~df['Discounted_Order']]['Total Order Value'].mean()
df.to_csv('data.csv', index=False)
```

```python
# Add calculated features to DataFrame
df['Discount_Ratio'] = discount_ratio
df['Avg_Discount_Percentage'] = avg_discount_percentage
df['Avg_Order_Value_Discounted'] = avg_order_value_discounted
df['Avg_Order_Value_Non_Discounted'] = avg_order_value_non_discounted
df.to_csv('data.csv', index=False)
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Calculate total sales for each product
df['Total_Sales'] = df['Discount Price'] * df['Total Order Value']

# Aggregate sales data by Product ID, Product Name, and Category
product_performance = df.groupby(['Product ID', 'Product Name', 'Category']).agg(
    total_sales=('Total_Sales', 'sum'),
    order_count=('Order ID', 'count')
).reset_index()

# Sort products by total sales
product_performance_sorted = product_performance.sort_values(by='total_sales', ascending=False)

df.to_csv('data.csv', index=False)

# Display top 10 most popular products
print(product_performance_sorted.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Calculate average sales for discounted and non-discounted products
discounted_sales = df[df['Discount Price'] > 0].groupby('Product ID')['Total_Sales'].mean()
non_discounted_sales = df[df['Discount Price'] == 0].groupby('Product ID')['Total_Sales'].mean()

# Combine into a single DataFrame for comparison
sales_comparison = pd.DataFrame({
    'Discounted_Avg_Sales': discounted_sales,
    'Non_Discounted_Avg_Sales': non_discounted_sales
}).reset_index()

# Plot the relationship between discount price and sales
sns.scatterplot(x=df['Discount Price'], y=df['Total_Sales'])
plt.title('Discount Price vs. Total Sales')
plt.xlabel('Discount Price')
plt.ylabel('Total Sales')
plt.show()

# Perform linear regression to quantify the relationship
slope, intercept, r_value, p_value, std_err = linregress(df['Discount Price'], df['Total_Sales'])
print(f"Linear Regression Results: Slope = {slope}, Intercept = {intercept}, R-squared = {r_value**2}")
df.to_csv('data.csv', index=False)
```

```python
# Add a column to identify whether a product was discounted
df['Is_Discounted'] = df['Discount Price'] > 0

# Calculate average sales for discounted and non-discounted products
discounted_sales = df[df['Is_Discounted']].groupby('Product ID')['Total_Sales'].mean()
non_discounted_sales = df[~df['Is_Discounted']].groupby('Product ID')['Total_Sales'].mean()

# Combine into a single DataFrame for comparison
sales_comparison = pd.DataFrame({
    'Discounted_Avg_Sales': discounted_sales,
    'Non_Discounted_Avg_Sales': non_discounted_sales
}).reset_index()

# Display the comparison
print(sales_comparison.head())
df.to_csv('data.csv', index=False)
```

```python
# Check for missing values in relevant columns
missing_data = df[['Discount Price', 'Total Order Value', 'Discount Percentage']].isna().sum()
print(missing_data)

# Handle missing values if necessary, e.g., fill with mean or median
df['Discount Price'].fillna(df['Discount Price'].median(), inplace=True)
df['Total Order Value'].fillna(df['Total Order Value'].median(), inplace=True)
df['Discount Percentage'].fillna(df['Discount Percentage'].median(), inplace=True)
df.to_csv('data.csv', index=False)
```

```python
# Create a column for discounted sales
df['Discounted_Sales'] = df['Total Order Value'] * (df['Discount Price'] / df['MRP'])

# Calculate average sales with and without discounts
avg_sales_discounted = df.groupby('Product ID')['Discounted_Sales'].mean()
avg_sales_non_discounted = df[df['Discount Price'] == 0].groupby('Product ID')['Total Order Value'].mean()

# Combine results into a single DataFrame
discount_effectiveness = pd.DataFrame({
    'Discounted_Avg_Sales': avg_sales_discounted,
    'Non_Discounted_Avg_Sales': avg_sales_non_discounted
}).reset_index()
df.to_csv('data.csv', index=False)
```

```python
from scipy.stats import ttest_ind

# Separate data for discounted and non-discounted sales
discounted_sales = df[df['Discount Price'] > 0]['Total Order Value']
non_discounted_sales = df[df['Discount Price'] == 0]['Total Order Value']

# T-test to compare means
t_stat, p_value = ttest_ind(discounted_sales, non_discounted_sales, equal_var=False)
print(f"T-test results: T-statistic = {t_stat}, P-value = {p_value}")
df.to_csv('data.csv', index=False)
```
